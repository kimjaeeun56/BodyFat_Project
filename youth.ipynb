# 9. 이상치 처리

```python
teen_df4.describe()
```

```python
# TEST_AGE 나이: (13~18)
# ITEM_F001 신장: (140이상)
# ITEM_F002 몸무게: (20이상)
# ITEM_F003 체지방율: (5이상 50이하)
# ITEM_F005 혈압(최저): 50이상 100이하 
# ITEM_F006 혈압(최고): 80이상 160이하 
# ITEM_F012 유연성: -50 ~ 50
# ITEM_F013 일리노이: 10~100(100이상 한명이라 )
# ITEM_F014 체공시간: 1~3
# ITEM_F017 협응력: 20-100
# ITEM_F018 BMI:10-50
# ITEM_F020 왕복오래달리기: 1이상 130이하 
# ITEM_F028 상대악력:10-100
```

```python
# 나이 (13~18)
age_idx = teen_df4[(13 > teen_df4['TEST_AGE']) | (teen_df4['TEST_AGE'] > 18)].index
teen_df4.drop(age_idx,inplace=True)
print(teen_df4.shape)
# 신장 (140 이상)
h_idx = teen_df4[140 > teen_df4['ITEM_F001']].index
teen_df4.drop(h_idx,inplace=True)
print(teen_df4.shape)
# 몸무게 (20이상 120이하)
w_idx = teen_df4[(20 > teen_df4['ITEM_F002']) | (teen_df4['ITEM_F002'] > 120)].index
teen_df4.drop(w_idx,inplace=True)
print(teen_df4.shape)
# 체지방율 (5이상 50이하)
r_idx = teen_df4[(5 > teen_df4['ITEM_F003']) | (teen_df4['ITEM_F003'] > 50)].index
teen_df4.drop(r_idx,inplace=True)
print(teen_df4.shape)
# 이완기혈압 (50이상 100이하)
b_idx = teen_df4[(50 > teen_df4['ITEM_F005']) | (teen_df4['ITEM_F005'] > 100)].index
teen_df4.drop(b_idx,inplace=True)
print(teen_df4.shape)
# 수축기혈압 (80이상 160이하)
p_idx = teen_df4[(80 > teen_df4['ITEM_F006']) | (teen_df4['ITEM_F006'] > 160)].index
teen_df4.drop(p_idx,inplace=True)
print(teen_df4.shape)
# 유연성
f_idx = teen_df4[(-50 > teen_df4['ITEM_F012']) | (teen_df4['ITEM_F012'] > 50)].index
teen_df4.drop(f_idx,inplace=True)
print(teen_df4.shape)
# 일리노이
i_idx = teen_df4[(10 > teen_df4['ITEM_F013']) | (teen_df4['ITEM_F013'] > 100)].index
teen_df4.drop(i_idx,inplace=True)
print(teen_df4.shape)
# 체공시간
a_idx = teen_df4[(0 > teen_df4['ITEM_F014']) | (teen_df4['ITEM_F014'] > 3)].index
teen_df4.drop(a_idx,inplace=True)
print(teen_df4.shape)
# 협응력
cor_idx = teen_df4[(20 > teen_df4['ITEM_F017']) | (teen_df4['ITEM_F017'] > 100)].index
teen_df4.drop(cor_idx,inplace=True)
print(teen_df4.shape)
# BMI
bmi_idx = teen_df4[(10 > teen_df4['ITEM_F018']) | (teen_df4['ITEM_F018'] > 50)].index
teen_df4.drop(bmi_idx,inplace=True)
print(teen_df4.shape)
# 심폐지구력(20m 왕복오래달리기)
sr_idx = teen_df4[(0 >= teen_df4['ITEM_F020']) | (teen_df4['ITEM_F020'] > 130)].index
teen_df4.drop(sr_idx,inplace=True)
print(teen_df4.shape)
# 근력(상대악력)
m_idx = teen_df4[(0 > teen_df4['ITEM_F028']) | (teen_df4['ITEM_F028'] > 100)].index
teen_df4.drop(m_idx,inplace=True)
print(teen_df4.shape)
# 심폐지구력(트레이드밀) 20-70?
tr_idx = teen_df4[(20 > teen_df4['ITEM_F035']) | (teen_df4['ITEM_F035'] > 70)].index
teen_df4.drop(tr_idx,inplace=True)
print(teen_df4.shape)
# 심폐지구력 (스텝)
st_idx = teen_df4[(20 > teen_df4['ITEM_F037']) | (teen_df4['ITEM_F037'] > 70)].index
teen_df4.drop(st_idx,inplace=True)
print(teen_df4.shape)
```

# 11. 심폐지구력

```python
shuttle_dict = {}

for i in range(shuttle_run.shape[0]):
    age = shuttle_run.iloc[i,0]
    a = str(shuttle_run.iloc[i,1])
    b = shuttle_run.iloc[i, 2].split('-')[0]
    c = shuttle_run.iloc[i, 2].split('-')[1]
    d = str(shuttle_run.iloc[i,3])
    e = str(shuttle_run.iloc[i,4])
    f = shuttle_run.iloc[i, 5].split('-')[0]
    g = shuttle_run.iloc[i, 5].split('-')[1]
    h = str(shuttle_run.iloc[i,6])
    shuttle_dict[age] = {0: (float(a), float(b), float(c), float(d)), 1: (float(e), float(f), float(g), float(h))}

shuttle_dict[13]
```

```python
ts_dict = {}

for i in range(t_s.shape[0]):
    age = t_s.iloc[i,0]
    a = str(t_s.iloc[i,1])
    b = t_s.iloc[i, 2].split('-')[0]
    c = t_s.iloc[i, 2].split('-')[1]
    d = str(t_s.iloc[i,3])
    e = str(t_s.iloc[i,4])
    f = t_s.iloc[i, 5].split('-')[0]
    g = t_s.iloc[i, 5].split('-')[1]
    h = str(t_s.iloc[i,6])
    ts_dict[age] = {0: (float(a), float(b), float(c), float(d)), 1: (float(e), float(f), float(g), float(h))}

ts_dict[13]
```

```python
teen_df4 = teen_df4.reset_index(drop=True)
teen_df4
```

```python
# 위험: 0, 개선: 1, 건강: 2
```

```python
'''
def shuttlerun_grade(age, sex, run_score):
    a, b, c, d = shuttle_dict[age][sex]
    if run_score <= a:
        return 0
    elif b <= run_score <= c:
        return 1
    elif run_score >= d:
        return 2
    else:
        return '이상치'
'''

def treadmill_grade(age, sex, t_score):
    a, b, c, d = ts_dict[age][sex]
    m_df = teen_df4[(teen_df4['TEST_AGE'] == age)&(teen_df4['TEST_SEX'] == 0)]
    f_df = teen_df4[(teen_df4['TEST_AGE'] == age)&(teen_df4['TEST_SEX'] == 1)]
    aa = m_df[m_df['ITEM_F020'] <= a]
    bc = m_df[(b <= m_df['ITEM_F020']) & (m_df['ITEM_F020'] <= c)]
    dd = m_df[m_df['ITEM_F020'] >= d]
    if t_score <= a:
        return aa['ITEM_F020'].mean()
    elif b <= t_score <= c:
        return bc['ITEM_F020'].mean()
    elif t_score >= d:
        return dd['ITEM_F020'].mean()
    else:
        return '이상치'

def step_grade(age, sex, s_score):
    a, b, c, d = ts_dict[age][sex]
    m_df = teen_df4[(teen_df4['TEST_AGE'] == age)&(teen_df4['TEST_SEX'] == 0)]
    f_df = teen_df4[(teen_df4['TEST_AGE'] == age)&(teen_df4['TEST_SEX'] == 1)]
    aa = m_df[m_df['ITEM_F020'] <= a]
    bc = m_df[(b <= m_df['ITEM_F020']) & (m_df['ITEM_F020'] <= c)]
    dd = m_df[m_df['ITEM_F020'] >= d]
    if s_score <= a:
        return aa['ITEM_F020'].mean()
    elif b <= s_score <= c:
        return bc['ITEM_F020'].mean()
    elif s_score >= d:
        return dd['ITEM_F020'].mean()
    else:
        return '이상치'
```

```python
심폐지구력 = []
check_df = teen_df4.isnull()

for i in teen_df4.index:
    age = teen_df4.loc[i, 'TEST_AGE']
    sex = teen_df4.loc[i, 'TEST_SEX']
    run_score = teen_df4.loc[i, 'ITEM_F020']
    t_score = teen_df4.loc[i, 'ITEM_F035'] 
    s_score = teen_df4.loc[i, 'ITEM_F037']

    if check_df.loc[i, 'ITEM_F020'] != True:
        심폐지구력.append(run_score)
    elif check_df.loc[i, 'ITEM_F037'] != True:
        심폐지구력.append(step_grade(age, sex, s_score))
    else: 
        심폐지구력.append(treadmill_grade(age, sex, t_score))
```

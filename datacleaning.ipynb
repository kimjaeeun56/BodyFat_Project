```python
shuttle_dict = {}

for i in range(shuttle_run.shape[0]):
    age = shuttle_run.iloc[i,0]
    a = str(shuttle_run.iloc[i,1])
    b = shuttle_run.iloc[i, 2].split('-')[0]
    c = shuttle_run.iloc[i, 2].split('-')[1]
    d = str(shuttle_run.iloc[i,3])
    e = str(shuttle_run.iloc[i,4])
    f = shuttle_run.iloc[i, 5].split('-')[0]
    g = shuttle_run.iloc[i, 5].split('-')[1]
    h = str(shuttle_run.iloc[i,6])
    shuttle_dict[age] = {0: (float(a), float(b), float(c), float(d)), 1: (float(e), float(f), float(g), float(h))}

shuttle_dict[13]
```

```python
ts_dict = {}

for i in range(t_s.shape[0]):
    age = t_s.iloc[i,0]
    a = str(t_s.iloc[i,1])
    b = t_s.iloc[i, 2].split('-')[0]
    c = t_s.iloc[i, 2].split('-')[1]
    d = str(t_s.iloc[i,3])
    e = str(t_s.iloc[i,4])
    f = t_s.iloc[i, 5].split('-')[0]
    g = t_s.iloc[i, 5].split('-')[1]
    h = str(t_s.iloc[i,6])
    ts_dict[age] = {0: (float(a), float(b), float(c), float(d)), 1: (float(e), float(f), float(g), float(h))}

ts_dict[13]
```

```python
strength_dict = {}

for i in range(strength.shape[0]):
    age = strength.iloc[i,0]
    a = str(strength.iloc[i,1])
    b = strength.iloc[i, 2].split('-')[0]
    c = strength.iloc[i, 2].split('-')[1]
    d = str(strength.iloc[i,3])
    e = str(strength.iloc[i,4])
    f = strength.iloc[i, 5].split('-')[0]
    g = strength.iloc[i, 5].split('-')[1]
    h = str(strength.iloc[i,6])
    strength_dict[age] = {0: (float(a), float(b), float(c), float(d)), 1: (float(e), float(f), float(g), float(h))}

strength_dict[13]
```

```python
# 위험: 0, 개선: 1, 건강: 2
```

```python
def shuttlerun_grade(age, sex, run_score):
    a, b, c, d = shuttle_dict[age][sex]
    if run_score <= a:
        return '위험'
    elif b <= run_score <= c:
        return '개선'
    elif run_score >= d:
        return '건강'
    else:
        return '이상치'

def treadmill_grade(age, sex, t_score):
    a, b, c, d = ts_dict[age][sex]
    if t_score <= a:
        return '위험'
    elif b <= t_score <= c:
        return '개선'
    elif t_score >= d:
        return '건강'
    else:
        return '이상치'

def step_grade(age, sex, s_score):
    a, b, c, d = ts_dict[age][sex]
    if s_score <= a:
        return '위험'
    elif b <= s_score <= c:
        return '개선'
    elif s_score >= d:
        return '건강'
    else:
        return '이상치'

def gripstrength_grade(age, sex, g_score):
    a, b, c, d = strength_dict[age][sex]
    if g_score <= a:
        return '위험'
    elif b <= g_score <= c:
        return '개선'
    elif g_score >= d:
        return '건강'
    else:
        return '이상치'
```

```python
심폐지구력_등급 = []
근력_등급 = []

for i in teen_df.index:
    age = teen_df.loc[i, '나이']
    sex = teen_df.loc[i, '성별']
    run_score = int(teen_df.loc[i, '심폐지구력'])
    g_score = int(teen_df.loc[i, '근력'])

    심폐지구력_등급.append(shuttlerun_grade(age, sex, run_score))
    근력_등급.append(gripstrength_grade(age, sex, g_score))
```

```python
teen_df['심폐지구력_등급'] = 심폐지구력_등급
teen_df['근력_등급'] = 근력_등급
```

```python
# bmi 등급 나누기

def bmi_grade(bmi):
    if bmi <= 18.5:
        return 0 # 저체중
    elif 18.5 < bmi <= 23.:
        return 1 # 정상
    elif 23. < bmi <= 25.:
        return 2 # 과체중
    elif 25. < bmi <= 35:
        return 3 # 비만
    else :
        return 4 # 고도비만
    # else:
    #     return 'who are you?'

def male_bodyfat_grade(bodyfat):
    if bodyfat < 10:
        return 0 # 저체중
    elif 10 <= bodyfat < 21.:
        return 1 # 정상
    elif 21 < bodyfat <= 25.:
        return 2 # 과체중
    elif 25. < bodyfat <= 35.:
        return 3 # 비만
    else :
        return 4 #고도비만
    # else:
    #     return 'who are you?'

def female_bodyfat_grade(bodyfat):
    if bodyfat < 18:
        return 0 # 저체중
    elif 18 <= bodyfat < 25.:
        return 1 # 정상
    elif 25 < bodyfat <= 30.:
        return 2 # 과체중
    elif 30. < bodyfat <= 40.:
        return 3 # 비만
    else:
        return 4 # 고도비만
    # else:
    #     return 'who are you?'
```

```python
teen_df['BMI_등급'] = teen_df['BMI'].apply(bmi_grade)
```
